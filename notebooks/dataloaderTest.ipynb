```python
total_duration_ms = segments[-1]['end'] * 1000 
```

```python
from pydub import AudioSegment
```

```python
final_audio = AudioSegment.silent(duration=total_duration_ms)
```

```python
from IPython.display import Audio
Audio(final_audio.raw_data, rate=final_audio.frame_rate)
```

```python
for segment in segments:
        # Use the audio data directly from the segment
        # The audio data is already loaded in the 'audio' field
        segment_audio = AudioSegment(
            data=segment['audio'].tobytes(),  # Convert numpy array to bytes
            sample_width=2,  # Correct: 16-bit audio is 2 bytes per sample
            frame_rate=segment['sampling_rate'],
            channels=1  # Assuming mono audio
        )
        
        # Print debug info
        print(f"Adding segment: duration={len(segment_audio)}ms, start_position={segment['start']*1000}ms")
        
        start_ms = segment['start'] * 1000
        final_audio = final_audio.overlay(segment_audio, position=int(start_ms))
```
